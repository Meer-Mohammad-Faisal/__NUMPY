# Advance operation with Business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

Data structure: [restaurant_id, 2021, 2022, 2023, 2024]

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Beijing Bites
    [3, 200000, 230000, 260000, 300000], #Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger king
    [5, 160000, 185000, 205000, 230000] # Chai Point
])
print("=== Zomato sales analysis ===")
print("\n sales data shape", sales_data.shape)
print("\n sample data for 1st 3 resturant: ", sales_data[:3])  # sales_data[:, 1:]) -- gives full data without id like(1,2,3,4,5)
```

# total sales per year

```python
# print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:], axis=0)
print(yearly_total)
```

```python
# total sum
final_sum = np.sum(yearly_total, axis=0)
print(final_sum)
```

minimum sales per resturant 


```python
min_sales = np.min(sales_data[:, 1:], axis=1) # axis 0 mtlb search in column wise and ( 0 means left to right)
                                              # axis 1 mtlb search in row wise and ( 1 means top to bottom)
print(min_sales)
```

maximum sales per year

```python
max_sales = np.max(sales_data[:, 1:], axis=0)
print(max_sales)
```

average sales per year

```python
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print(avg_sales)
```

# cumalative sales

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restarunt")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

# now some vector

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ", vector1 + vector2)
print("Vector multiplication: ", vector1 * vector2)
print("Dot product", np.dot(vector1, vector2))

# angle 
angle = np.arccos(np.dot(vector1, vector2)/ (np.linalg.norm(vector1)
* np.linalg.norm(vector2)))
print("angle is", angle)
```
